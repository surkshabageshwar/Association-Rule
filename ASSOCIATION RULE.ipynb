```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
basket=(data[data['Country']=='France'].groupby(['InvoiceNo','Description'])
        ['Quantity'].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo"))
```

```python
basket
```

```python
basket.drop("POSTAGE",1,inplace=True)
```

```python
basket
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
         return 1
basket=basket.applymap(encode_units)
```

```python
frq_item=apriori(basket,min_support=0.01,use_colnames=True)
```

```python
rule=association_rules(frq_item,min_threshold=1,metric='lift')
```

```python
len(rule)
```

```python
r1=rule.sort_values(('lift'),ascending=False)
```
